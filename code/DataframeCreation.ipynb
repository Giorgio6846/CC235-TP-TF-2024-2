```python
import pandas as pd
import os

faceDir = "../data/NU_Dataset_1/Images/FaceNorm"
depthDir = "../data/NU_Dataset_1/Images/DepthNorm"

depthFiles = []
faceFiles = []

for root, dirs, files in os.walk(depthDir):
    for file in files:
        if file.endswith("_4.jpg"):
            depthFiles.append(os.path.join(root, file))

for root, dirs, files in os.walk(faceDir):
    for file in files:
        if file.endswith("_4.jpg"):
            faceFiles.append(os.path.join(root, file))

faceFiles1 = []
faceFiles2 = []

depthFiles1= []
depthFiles2= []

for faceFile in faceFiles:
    if faceFile.endswith("RGB_A_1_4.jpg"):
        faceFiles1.append(faceFile)
    if faceFile.endswith("RGB_A_2_4.jpg"):
        faceFiles2.append(faceFile)
    if faceFile.endswith("RGB_A_3_4.jpg"):
        faceFiles1.append(faceFile)
    if faceFile.endswith("RGB_A_4_4.jpg"):
        faceFiles2.append(faceFile)

for depthFile in depthFiles:
    if depthFile.endswith("RGB_A_1_4.jpg"):
        depthFiles1.append(depthFile)
    if depthFile.endswith("RGB_A_2_4.jpg"):
        depthFiles2.append(depthFile)
    if depthFile.endswith("RGB_A_3_4.jpg"):
        depthFiles1.append(depthFile)
    if depthFile.endswith("RGB_A_4_4.jpg"):
        depthFiles2.append(depthFile)


dataframe = pd.DataFrame(
    list(zip(faceFiles1, depthFiles1, faceFiles2, depthFiles2)), columns=["Face1", "Depth1", "Face2", "Depth2"]
)

dataframe.loc[len(dataframe.index)] = [
    "../data/Dataset_2/User1/Test/FaceImage20241112204422.jpg",
    "../data/Dataset_2/User1/Test/DepthImage20241112204422.png",
    "../data/Dataset_2/User1/Validation/FaceImage20241112204140.jpg",
    "../data/Dataset_2/User1/Validation/DepthImage20241112204140.png"
]

dataframe["Validation"] = 1
print(dataframe.head())

dataframe.to_csv(
    "../data/data.csv", index=False
)
```
