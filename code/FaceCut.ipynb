# Modificaion dataset

```python
import cv2
import matplotlib.pyplot as plt
import os

def detect_faces(image, depth):
    face_cascade = cv2.CascadeClassifier(
        cv2.data.haarcascades + "haarcascade_frontalface_alt2.xml"
    )
    faces = face_cascade.detectMultiScale(
        image, scaleFactor=1.19, minNeighbors=15, minSize=(80, 80)
    )
    if len(faces) == 0:
        return None
    x, y, w, h = faces[0]

    face_image = image[y : y + h, x : x + w]
    depth_image = depth[y : y + h, x : x + w]

    face_image_resized = cv2.resize(face_image, (480, 480))
    depth_image_resized = cv2.resize(depth_image, (480, 480))

    return face_image_resized, depth_image_resized
```

## Recorte de imagenes

```python
faceDir = "../data/NU_Dataset_1/Images/Face"
depthDir = "../data/NU_Dataset_1/Images/Depth"

depthFiles = []
faceFiles = []

for root, dirs, files in os.walk(depthDir):
    for file in files:
        if file.endswith("_4.jpg"):
            depthFiles.append(os.path.join(root, file))

for root, dirs, files in os.walk(faceDir):
    for file in files:
        if file.endswith("_4.jpg"):
            faceFiles.append(os.path.join(root, file))

print(faceFiles)
print(depthFiles)

print(os.path.basename(faceFiles[0]))
```

```python
faceNormDir = "../data/NU_Dataset_1/Images/FaceNorm"
depthNormDir = "../data/NU_Dataset_1/Images/DepthNorm"

for faceFile in faceFiles:
    faceIMGs = []
    depthIMGs = []

    fileName = os.path.basename(faceFile)

    faceFile  = f"{faceDir}/{fileName}"
    depthFile = f"{depthDir}/{fileName}"

    faceIMG = cv2.imread(faceFile)[..., ::-1]
    depthIMG = cv2.imread(depthFile)

    images = detect_faces(faceIMG, depthIMG)

    if images != None:
        faceRGB = images[0][..., ::-1]

        cv2.imwrite(f"{faceNormDir}/{fileName}", faceRGB)
        cv2.imwrite(f"{depthNormDir}/{fileName}", images[1])
```
